### Student Management System
##### Objective: Create a system that helps manage student records by adding, updating, and deletinstudent 

```python
# Student Management System

class Student:
    def __init__(self, student_ID):
        self.name = input("Name:  ")
        self.name = self.name.upper()
        self.age = int(input("Age:  "))
        self.grade = input("Grade:  ")
        self.grade = self.grade.upper()
        self.ID = student_ID

    def __str__(self):
        return f"Student_ID: {self.ID}, Name: {self.name}, Age: {self.age}, Grade: {self.grade}"

# Function to add a student into the system
def Add(student):
        students.append(student)

# Function to update a student's information
def Update(student_ID):
    for stud in students:
        if stud.ID == student_ID:
            
            print("\n")
            print(stud,"\n")

            print("Add new data to edit a field or leave blank to maintain old data\n")

            print("Name")
            new_name = input("Enter new name:  ")
            if len(new_name) != 0:
                stud.name = new_name.upper()
                print("Name UPDATED!\n")

            print("\nAge")
            new_age = input("Enter new age:  ")
            if len(new_age) != 0:
                stud.age = int(new_age)
                print("Age UPDATED!")

            print("\nGrade")
            new_grade = input("Enter new garde: ")
            if len(new_grade) == 1:
                stud.grade = new_grade.upper()
                print("Grade UPDATED!")

# Function to view all students in our system, sorted alphabetically, in a tabular display
def ViewAll(students):
    print("|","ID"," "*2,"|", "Full Name"," "*(20),"|", "Age","|", "Grade","|")
    print("|-------|--------------------------------|-----|-------|")

    students.sort(key = lambda stud: stud.name)
    for stud in students:
        print("|"+stud.ID,"|", stud.name," "*(29-len(stud.name)),"|", stud.age," |", stud.grade+"     |")
        print("|-------|--------------------------------|-----|-------|")

# Function to delete a student from our system
def Delete(student_ID):
    for stud in students:
        #for stud_info in stud:
        if stud.ID == student_ID:
            print("This student information will be DELETED! \n",stud)
            confirm = int(input("\n1. To confirm your DELETE! \t\t 0. To cancel\n"))
            if confirm == 1:
                students.remove(stud)
                print("DELETED!!!")
                break
            else:
                break

# Function to check if a student ID is valid according to our defined format
def ValidID(student_ID):
    if len(student_ID) == 6 and student_ID[2] == 'P' and int(student_ID[0]+student_ID[1]+student_ID[3]+student_ID[4]+student_ID[5]) in range(25001, 25201):
        return True
    else:
        return False

# Function to check if an ID is alreddy registered in the system
def CheckID(student_ID):
    if ValidID(student_ID):
        for stud in students:
            if student_ID == stud.ID:
                return True
        return False
    else:
        return False

# Fuction to display our main menu
def Homepage():
    print("\n\nWELCOME TO THE STUDENT MANAGEMENT PLATFORM\n")
    print("1) Add Student\n")
    print("2) Update Student\n")
    print("3) View All Students\n")
    print("4) Delete Student\n")
    print("0) Exit\n")
    menu_number = input("Enter either 1, 2, 3, 4 or 0 for the different options.\n")
    return menu_number

print("\n\nNOTE: A VAILD ID IS OF THE FORMAT 25P158, AND IN THE RANGE 25P001 TO 25P200")

students = [] # used to store all students information

menu_number = Homepage()

while menu_number not in ['1', '2', '3', '4', '0']:
    menu_number = input()

while menu_number in ['1', '2', '3', '4', '0']:

    if menu_number == '1':    # Menu to add a student into the system
        print("\nAdd Student")
        flag = '1'
        while flag == '1':
            student_ID = input("\nEnter student's ID:  ")
            student_ID = student_ID.replace(student_ID[2], student_ID[2].upper())
            if ValidID(student_ID):
                if CheckID(student_ID):
                    print("This ID is already taken!\n")
                    flag = '1'
                else:
                    student = Student(student_ID)
                    Add(student)
                    flag = input("\n1. Add another student \t\t 0. Save and Go Back to Homepage\n")
                    while flag != '0' and flag != '1':
                        flag = input()
            else:
                print("Invalid ID format!\nEnter a valid ID\n")
                flag = '1'

    if menu_number == '2':    # Menu to update student information
        if students == []:
            print("\nYou don't have any student in your list!")
        else:
            print("\nUpdate Student")
            flag = '1'
            while flag == '1':
                student_ID = input("\nEnter student's ID:  ")
                student_ID = student_ID.replace(student_ID[2], student_ID[2].upper())
                if CheckID(student_ID):
                    Update(student_ID)
                    flag = input("\n1. Update another student \t\t 0. Save and Go Back to Homepage\n")
                    while flag != '0' and flag != '1':
                        flag = input()
                    if flag == '0':
                        break
                else:
                    print("This ID does not exist!\n")
                    flag = '1'

    if menu_number == '3':    # Menu to view all students registered in our system
        if students == []:
            print("\nYou don't have any student in your list!\n")
        else:
            print("\nLIST OF STUDENTS IN YOUR SYSTEM\n")
            ViewAll(students)
            flag = input("\n0. Go Back to Homepage\n")
            while flag != '0':
                flag = input()
    
    if menu_number == '4':    # Menu to delete a student from our system
        if students == []:
            print("\nYou don't have any student in your list!")
        else:
            print("\nDelete Student")
            flag = '1'
            while flag == '1':
                student_ID = input("\nEnter student's ID:  ")
                student_ID = student_ID.replace(student_ID[2], student_ID[2].upper())
                if CheckID(student_ID):
                    Delete(student_ID)
                    flag = input("\n1. Delete another student \t\t 0. Go Back to Homepage\n")
                    while flag != '0' and flag != '1':
                        flag = input()
                else:
                    print("This ID does not exist!\n")
                    flag = '1'

    if menu_number == '0':
        break

    menu_number = Homepage()
    while menu_number not in ['1', '2', '3', '4', '0']:
        menu_number = input()
```

### Mini calculator
##### Objective. Develop a calculator that performs basic arithmetic operations.

```python
# Mini Calculator App

def product(a ,b):
    return a * b

def divide(a, b):
    return a/b

def subtract(a, b):
    return a - b

def add(a, b):
    return a + b

flag = '1'
while flag == '1':

    try:
        a = float(input("\nEnter first operand: "))
        operator = input("Enter operator: ")
        b = float(input("Enter second operand: "))

        if operator == '+':
            result = add(a, b)

        elif operator == '-':
            result = subtract(a, b)

        elif operator == '*':
            result = product(a, b)

        elif operator == '/':
            result = divide(a, b)

        else:
            raise Exception

    except ValueError:
        print("Operations can only be performed on numbers.")
    except ZeroDivisionError:
        print("You can't divide by 0.")
    except Exception:
        print("Your operator must be one of the following: (+, -, *, /)")
    else:
        print(f"Result: {result}")

    flag = input("\nEnter 1 to perform another calculation \t|\t Press any key to exit\n")
```
